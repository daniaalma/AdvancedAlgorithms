    * The sizes of the lists are at least two or more.
    * Both lists are sorted in ascending order.

    * Note that your algorithm, if correct, will always succeed in finding an intersection point.
    
Running time of your algorithm must be $\Theta(\log_2(n))$.


```python
def bs2(xValues, yValues, left, right):
    
    if right>=left:
        midp=(left+right)//2
        
        if yValues[midp] > xValues[midp] and yValues[midp+1] < xValues[midp+1]:
            return midp
            
        if yValues[midp] > xValues[midp]:
            return bs2(xValues,yValues,(midp+1),right)
        
        else:
            return bs2(xValues,yValues,left,(midp-1))
        
        
        
#so x=y is the line we are trying to find intersection with, but no points are x=y in our coordinate system...
#assumptions we can make for intersection:
#between the x=y curve intersection point (y is labeled y2)
#y at (xi,yi) is >= y2 and y at (xi+1,yi+1) is <= y2

#none of the y's are negative

def findPointOfIntersection(xValues, yValues): 
    return(bs2(xValues,yValues,0,len(xValues)))
```

__(A, 7 points)__  Consider the Fourier transform of the circular permutation $(a_1, a_2, \ldots, a_{n-1}, a_0)$. Let us denote it as $( \hat{A}_0, \ldots, \hat{A}_{n-1})$. Find an expression for $\hat{A}_j$ in terms of $A_j$. **Hint** Use the fact that $\omega_n^n = 1$.

__(B, 8 points)__ Generalize part (A) for the circular permutation $(a_{p}, \ldots, a_{n-1}, a_0, \ldots, a_{p-1})$. Use the fact that $\omega_n^{n+j} = \omega_n^j$.

### Answer (15 lines)

$$ 
\begin{align*}
A_j = \sum_{k=0}^{n-1} a_k (\omega_n^{jk})
\end{align*}
$$

where 
$\omega_n = e^{\left( \frac{2 \pi i }{n} \right)}$


# A

#beginning series (before circular permutation):

$(a_0, \ldots, a_{n-1}) =>FFT=> (A_0, A_1, \ldots, A_{n-1})$


#end series (after circular permutation):

$(a_1, a_2, \ldots, a_{n-1}, a_0) =>FFT=> \hat{A}_j$ in terms of $A_j$

$$
\begin{align*}
A_j&=a_0(\omega_n^j)^0 + a_1(\omega_n^j)^1 + ... +a_{n-1}(\omega_n^j)^{n-1}\\
\hat{A}_j&=a_1(\omega_n^j)^0 + a_2(\omega_n^j)^1 + ... +a_{n-1}(\omega_n^j)^{n-2}+a_0(\omega_n^j)^{n-1}
\end{align*}
$$


#so what can we do to turn $\hat{A}_j$ to $A_j$? 


$$
\begin{align*}
A_j&=a_0(\omega_n^j)^0 + a_1(\omega_n^j)^1 + ... +a_{n-1}(\omega_n^j)^{n-1}\\
(\omega_n^j)*(\hat{A}_j)&=(\omega_n^j)*(a_1(\omega_n^j)^0 + a_2(\omega_n^j)^1 + ... +a_{n-1}(\omega_n^j)^{n-2}+a_0(\omega_n^j)^{n-1})\\
\hat{A}_j&=a_1(\omega_n^j)^1 + a_2(\omega_n^j)^2 + ... +a_{n-1}(\omega_n^j)^{n-1}+a_0(\omega_n^j)^{n}
\end{align*}
$$

if $\omega_n^n = 1$

the term $a_0(\omega_n^j)^{n}$ would then become $a_0$


therefore:

$$
\begin{align*}
(\omega_n^j)*(\hat{A}_j)&=A_j
\end{align*}
$$

## Problem 4 (15 points)

Show that the summation 

$$ \sum_{j=0}^{n-1} \left( \sum_{k=0}^{n-1} a_k \omega^{jk}\right)  (\overline{\omega})^{j \color{red}{m}} = n a_{\color{red}{m}} $$
for $m = 0, \ldots, n-1$. Here $\omega$ is the $n^{th}$ root of unity and $\overline{\omega} = \omega^{-1}$.
This establishes the fact that the inverse of the FFT is just the same algorithm using $\overline{\omega}$ instead of $\omega$, as discussed in class.

**Hints** Do not panic!!  

1. Bring the $(\overline{\omega})^{j \color{red}{m}}$ term into the summation, using the fact that $\overline{\omega} = \omega^{-1}$.

2. Next, change the order of summation taking each $a_k$ out in common.

3. Use the geometric series summation and you will see all coefficients cancel to $0$ other than that for $a_m$.

4. Arrive at the answer.

## Solution (Expected Length: 15 lines)


Prove the following:

$$ 
\begin{align*}
\sum_{j=0}^{n-1} \left( \sum_{k=0}^{n-1} a_k \omega^{jk}\right)  (\overline{\omega})^{j \color{red}{m}} = n a_{\color{red}{m}}\\
\end{align*}
$$

Step 1 on DO NOT PANIC!!:

Bring $(\overline{\omega})^{j \color{red}{m}}$ in and $\overline{\omega} = \omega^{-1}$

$$ 
\begin{align*}
\sum_{j=0}^{n-1} \left( \sum_{k=0}^{n-1} a_k \omega^{jk}\right)  (\overline{\omega})^{j \color{red}{m}} &= \sum_{j=0}^{n-1} \left( \sum_{k=0}^{n-1} a_k \omega^{jk} (\omega^{-j \color{red}{m}})\right)\\
&= \sum_{j=0}^{n-1} \left( \sum_{k=0}^{n-1} a_k (\omega^{j( k-\color{red}{m})})\right)\\
\end{align*}
$$


Step 2 on DO NOT PANIC!!:

take out the $a_k$, switch the summations because its really just sum so you can do it either way

$$ 
\begin{align*}
\sum_{j=0}^{n-1} \left( \sum_{k=0}^{n-1} a_k \omega^{jk}\right)  (\overline{\omega})^{j \color{red}{m}} &= \sum_{k=0}^{n-1} a_k \left( \sum_{j=0}^{n-1} (\omega^{j( k-\color{red}{m})})\right)\\
\end{align*}
$$

Step 3 on DO NOT PANIC!!:
use geometric series summation on the second term to prove it'll always go to 0 except in one case.

$$ 
\begin{align*}
\sum_{j=0}^{n-1} \left( \sum_{k=0}^{n-1} a_k \omega^{jk}\right)  (\overline{\omega})^{j \color{red}{m}} &= \sum_{j=0}^{n-1} a_k \left( \sum_{k=0}^{n-1} (\omega^{j( k-\color{red}{m})})\right)\\
\end{align*}
$$

for the second term:

$$ 
\begin{align*}
\sum_{k=0}^{n-1} (\omega^{j( k-\color{red}{m})})&= \frac{\omega^{n(k-m)}-1}{\omega^{k-m}-1}\\
\end{align*}
$$

in any case where $k\neq m$ , we will get a 0, since $\omega^{n}=1$, nth root of unity:

$$ 
\begin{align*}
\frac{\omega^{n(k-m)}-1}{\omega^{k-m}-1}=\frac{1-1}{\omega^{k-m}-1}=0\\
\end{align*}
$$

in the one case that it doesn't cancel to 0, where k=m:

$$ 
\begin{align*}
\sum_{k=0}^{n-1} (\omega^{j( k-\color{red}{m})})&=\sum_{k=0}^{n-1} (\omega^{j( m-\color{red}{m})})\\
&=\sum_{k=0}^{n-1} (\omega^{0})\\
&=n
\end{align*}
$$

we put it all together at k=m

$$ 
\begin{align*}
\sum_{j=0}^{n-1} \left( \sum_{k=0}^{n-1} a_k \omega^{jk}\right)  (\overline{\omega})^{j \color{red}{m}} &= \sum_{k=m}^{n-1} a_k *n\\
&= n a_{\color{red}{m}}\\
\end{align*}
$$

Step 3 on DO NOT PANIC!!:
arrive at the answer

$$ 
\begin{align*}
\sum_{j=0}^{n-1} \left( \sum_{k=0}^{n-1} a_k \omega^{jk}\right)  (\overline{\omega})^{j \color{red}{m}} &= n a_{\color{red}{m}}\\
\end{align*}
$$
